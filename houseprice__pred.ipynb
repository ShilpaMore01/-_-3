```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/Housing.csv")
data
```

```python
x=data[['area','bedrooms','bathrooms']]
y=data['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=23)
```

```python
from sklearn.linear_model import LinearRegression
ols=LinearRegression()
ols.fit(x_train,y_train)
prediction=ols.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error as mse
print('OLS',mse(y_test,prediction))
print('OLS',np.sqrt(mse(y_test,prediction)))
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.linear_model import SGDRegressor
lr=SGDRegressor()
lr.fit(x_train,y_train)
prediction2=lr.predict(x_test)
```

```python
print('Linear Regression test',mse(y_test,prediction2))
print('Linear Regression train',mse(y_train,lr.predict(x_train)))
print('Linear Regression test',np.sqrt(mse(y_test,prediction2)))
```
